```python
def readable_num(num):
    num_string = str(num)
    result = ''
    for i in range(len(num_string)):
        result = result + num_string[i] 
        if (len(num_string)-1-i) % 3 == 0:
            result = result + ','
    result = result.rstrip(',')
    return result

def start_grid():
    numstring = '''08 02 22 97 38 15 00 40 00 75 04 05 07 78 52 12 50 77 91 08
    49 49 99 40 17 81 18 57 60 87 17 40 98 43 69 48 04 56 62 00
    81 49 31 73 55 79 14 29 93 71 40 67 53 88 30 03 49 13 36 65
    52 70 95 23 04 60 11 42 69 24 68 56 01 32 56 71 37 02 36 91
    22 31 16 71 51 67 63 89 41 92 36 54 22 40 40 28 66 33 13 80
    24 47 32 60 99 03 45 02 44 75 33 53 78 36 84 20 35 17 12 50
    32 98 81 28 64 23 67 10 26 38 40 67 59 54 70 66 18 38 64 70
    67 26 20 68 02 62 12 20 95 63 94 39 63 08 40 91 66 49 94 21
    24 55 58 05 66 73 99 26 97 17 78 78 96 83 14 88 34 89 63 72
    21 36 23 09 75 00 76 44 20 45 35 14 00 61 33 97 34 31 33 95
    78 17 53 28 22 75 31 67 15 94 03 80 04 62 16 14 09 53 56 92
    16 39 05 42 96 35 31 47 55 58 88 24 00 17 54 24 36 29 85 57
    86 56 00 48 35 71 89 07 05 44 44 37 44 60 21 58 51 54 17 58
    19 80 81 68 05 94 47 69 28 73 92 13 86 52 17 77 04 89 55 40
    04 52 08 83 97 35 99 16 07 97 57 32 16 26 26 79 33 27 98 66
    88 36 68 87 57 62 20 72 03 46 33 67 46 55 12 32 63 93 53 69
    04 42 16 73 38 25 39 11 24 94 72 18 08 46 29 32 40 62 76 36
    20 69 36 41 72 30 23 88 34 62 99 69 82 67 59 85 74 04 36 16
    20 73 35 29 78 31 90 01 74 31 49 71 48 86 81 16 23 57 05 54
    01 70 54 71 83 51 54 69 16 92 33 48 61 43 52 01 89 19 67 48'''

    # Convert the string of numbers into a two-dimensional list of integers (origin in the top left)
    numgrid = [x.split() for x in numstring.split('\n')]
    for i in range(len(numgrid)):
        for j in range(len(numgrid[i])):
            numgrid[i][j] = int(numgrid[i][j])
    
    return numgrid

def horizontal_solution(grid):
    horizontal_numbers = []
    horizontal_product = 0

    # Co-ordinates for starting values for each row of four values to be checked
    for y in range(20):
        for x in range(16):

            numbers_list = []
            product = 1

            # Creates the list of four horizontal values, increasing the x co-ordinate
            for c in range(4):
                numbers_list.append(grid[y][x+c])

            for num in numbers_list:
                product *= num

            if product > horizontal_product:
                horizontal_product = product
                horizontal_numbers = numbers_list[:]
    
    return horizontal_product

def vertical_solution(grid):
    vertical_numbers = []
    vertical_product = 0

    # Co-ordinates for starting values for each column of four values to be checked
    for x in range(20):
        for y in range(16):

            numbers_list = []
            product = 1

            # Creates the list of four vertial values, increasing the y co-ordinate (going down)
            for c in range(4):
                numbers_list.append(grid[y+c][x])

            for num in numbers_list:
                product *= num

            if product > vertical_product:
                vertical_product = product
                vertical_numbers = numbers_list[:]

    return vertical_product

def diagonal_solution(grid):
    diagonal_numbers = []
    diagonal_product = 0

    # Co-ordinates for starting values for each right-down diagonal four values to be checked 
    for y in range(16):
        for x in range(16):

            numbers_list = []
            product = 1

            # Creates the list of four right-down diagonal values, increasing both the x and y co-ordinates
            for c in range(4):
                numbers_list.append(grid[y+c][x+c])
            
            for num in numbers_list:
                product *= num

            if product > diagonal_product:
                diagonal_product = product
                diagonal_numbers = numbers_list[:]

    # Co-ordinates for starting values for each left-down diagonal four values to be checked
    for y in range(16):
        for x in range(3, 19):

            numbers_list = []
            product = 1

            # Creates the list of four left-down diagonal values, increasing the y co-ordinate, but decreasng the x
            for c in range(4):
                numbers_list.append(grid[y+c][x-c])
            
            for num in numbers_list:
                product *= num

            if product > diagonal_product:
                diagonal_product = product
                diagonal_numbers = numbers_list[:]

    return diagonal_product

print('The greatest product of four adjacent numbers horizontally in the 20 by 20 grid:', readable_num(horizontal_solution(start_grid())))
print('The greatest product of four adjacent numbers vertically in the 20 by 20 grid:', readable_num(vertical_solution(start_grid())))
print('The greatest product of four adjacent numbers diagonally in the 20 by 20 grid:', readable_num(diagonal_solution(start_grid())))
```

## Problem 12: Highly Divisible Triangular Number <a id="problem12"></a>

The sequence of triangle numbers is generated by adding the natural numbers. So the 7<sup>th</sup> triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

1: 1\
3: 1, 3\
6: 1, 2, 3, 6\
10: 1, 2, 5, 10\
15: 1, 3, 5, 15\
21: 1, 3, 7, 21\
28: 1, 2, 4, 7, 14, 28

We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

### Solution

-

```python
def readable_num(num):
    num_string = str(num)
    result = ''
    for i in range(len(num_string)):
        result = result + num_string[i] 
        if (len(num_string)-1-i) % 3 == 0:
            result = result + ','
    result = result.rstrip(',')
    return result

def solution():
    triangle_num = 1
    natural_num = 1
    number_of_factors = 1

    while number_of_factors < 500:
        natural_num += 1
        triangle_num += natural_num

        num = triangle_num
        number_of_factors = 1
        factor = 2

        while factor <= num:
            exp = 1

            while num % factor == 0:
                num /= factor
                exp += 1
            
            number_of_factors *= exp
            factor += 1

        if num == triangle_num or num > 1:
            number_of_factors *= 2

    return triangle_num

print('The value of the first triangle number to have over five hundred divisors:', readable_num(solution()))
```

## Problem 14: Longest Collatz Sequence <a id="problem14"></a>

The following iterative sequence is defined for the set of positive integers:

n → n/2 (n is even)\
n → 3n + 1 (n is odd)

Using the rule above and starting with 13, we generate the following sequence:

13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1

It can be seen that this sequence (starting at 13 and finishing at 1) contains 10 terms.\
Although it has not been proved yet (Collatz Problem), it is thought that all starting numbers finish at 1.

Which starting number, under one million, produces the longest chain?

NOTE: Once the chain starts the terms are allowed to go above one million.

### Solution

-

```python
def solution():
    most_terms = 1
    max_n = 1
    chains = {}

    for _n in range(1, 1_000_000):
        n = _n
        terms = 1
        while n != 1:
            if n in chains.keys():
                terms += chains[n] - 1
                break
            elif n % 2 == 0:
                n /= 2
            else:
                n = (3*n) + 1
            terms += 1
        if _n not in chains.keys():
            chains[_n] = terms
        if terms > most_terms:
            most_terms = terms
            max_n = _n

    return max_n

print('The starting number under one million that produces the longest chain:', solution())
```

## Problem 15: Lattice Paths <a id="problem15"></a>

Starting in the top left corner of a 2 $\times$ 2 grid, and only being able to move to the right and down, 
there are exactly 6 routes to the bottom right corner.

How many such routes are there through a 20 $\times$ 20 grid?

### Solution

This is simply the centreal binomial coefficient for n = 20: $40 \choose 20$.

This is because all routes are exactly 40 moves that contain an equal number of 'down' and 'right' moves.

$40 \choose 20$ gives the total number of ways 20 options can be chosen from 40 moves and is derived by $\frac{40!}{20!^2}$.

```python
def readable_num(num):
    num_string = str(num)
    result = ''
    for i in range(len(num_string)):
        result = result + num_string[i] 
        if (len(num_string)-1-i) % 3 == 0:
            result = result + ','
    result = result.rstrip(',')
    return result

def solution():
    prod = 1
    for i in range(1, 21):
        prod *= (20+i)/i

    return round(prod)

print('The number of routes through a 20 by 20 grid are:', readable_num(solution()))
```

## Problem 17: Number Letter Counts <a id="problem17"></a>

If the numbers 1 to 5 are written out in words: one, two, three, four, five, then there are 3 + 3 + 5 + 4 + 4 = 19 letters used in total.

If all the numbers from 1 to 1000 (one thousand) inclusive were written out in words, how many letters would be used?

NOTE: Do not count spaces or hyphens. For example, 342 (three hundred and forty-two) contains 23 letters and 115 (one hundred and fifteen) contains 20 letters.\
The use of "and" when writing out numbers is in compliance with British usage.

### Solution

-

```python
def solution():
    
    units = {0:"", 1:"one", 2:"two", 3:"three", 4:"four", 5:"five", 6:"six", 7:"seven", 8:"eight", 9:"nine", 10:"ten", \
        11:"eleven", 12:"twelve", 13:"thirteen", 14:"fourteen", 15:"fifteen", 16:"sixteen", 17:"seventeen", 18:"eighteen", 19:"nineteen"}
    tens = {2:"twenty", 3:"thirty", 4:"forty", 5:"fifty", 6:"sixty", 7:"seventy", 8:"eighty", 9:"ninety"}

    total = 0
    for i in range (1, 1001):
        num = i
        num_string = ''

        if num == 1000:
            num_string = units[num//1000] + "thousand"
            num -= ((num//1000)*1000)

        if num > 99:
            num_string += units[num//100] + "hundred"
            num -= ((num//100)*100)

        if num != 0 and num_string != '':
            num_string += "and"

        if num > 19 :
            num_string += tens[num//10]
            num -= ((num//10)*10)

        num_string += units[num]
        total += len(num_string)

    return total

print('The number of letters used if all the numbers from 1 to 1000 inclusive were written out in words:', solution())
```

## Problem 18: Maximum Path Sum <a id="problem18"></a>

By starting at the top of the triangle below and moving to adjacent numbers on the row below, the maximum total from top to bottom is 23.

       3
      7 4
     2 4 6
    8 5 9 3

That is, 3 + 7 + 4 + 9 = 23.

Find the maximum total from top to bottom of the triangle below:

                         75
                       95 64
                      17 47 82
                    18 35 87 10
                  20 04 82 47 65
                 19 01 23 75 03 34
                88 02 77 73 07 63 67
              99 65 04 28 06 16 70 92
            41 41 26 56 83 40 80 70 33
           41 48 72 33 47 32 37 16 94 29
          53 71 44 65 25 43 91 52 97 51 14
         70 11 33 28 77 73 17 78 39 68 17 57
       91 71 52 38 17 14 91 43 58 50 27 29 48
      63 66 04 68 89 53 67 30 73 16 69 87 40 31
    04 62 98 27 23 09 70 98 73 93 38 53 60 04 23

NOTE: As there are only 16384 routes, it is possible to solve this problem by trying every route.\
However, Problem 67, is the same challenge with a triangle containing one-hundred rows; it cannot be solved by brute force, and requires a clever method!

### Solution

-

```python
def solution():
    triangle = '''75
                95 64
                17 47 82
            18 35 87 10
            20 04 82 47 65
            19 01 23 75 03 34
            88 02 77 73 07 63 67
        99 65 04 28 06 16 70 92
        41 41 26 56 83 40 80 70 33
        41 48 72 33 47 32 37 16 94 29
        53 71 44 65 25 43 91 52 97 51 14
    70 11 33 28 77 73 17 78 39 68 17 57
    91 71 52 38 17 14 91 43 58 50 27 29 48
    63 66 04 68 89 53 67 30 73 16 69 87 40 31
    04 62 98 27 23 09 70 98 73 93 38 53 60 04 23'''

    # Converting the triangle into a collection of lists of integers
    t = [x.split() for x in triangle.split("\n")]
    for i in range(len(t)):
        for j in range(len(t[i])):
            t[i][j] = int(t[i][j])

    # Bottom-up dynamic programming, calculating the sum of each number with the max of its adjacents below
    for i in range(len(t)-2, -1, -1):
        for j in range(i+1):
            t[i][j] += max(t[i+1][j], t[i+1][j+1])
    return t[0][0]

print('The maximum total from top to bottom of the given triangle:', solution())
```

## Problem 19: Counting Sundays <a id="problem19"></a>

You are given the following information, but you may prefer to do some research for yourself.

1 Jan 1900 was a Monday.\
Thirty days has September,\
April, June and November.\
All the rest have thirty-one,\
Saving February alone,\
Which has twenty-eight, rain or shine.\
And on leap years, twenty-nine.

A leap year occurs on any year evenly divisible by 4, but not on a century unless it is divisible by 400.

How many Sundays fell on the first of the month during the twentieth century (1 Jan 1901 to 31 Dec 2000)?

### Solution

-

```python
def solution():
    days_in_month = {1:31, 2:28, 3:31, 4:30, 5:31, 6:30, 7:31, 8:31, 9:30, 10:31, 11:30, 12:31}

    day, month, year = 1, 1, 1900
    day_of_week = 1
    counter = 0

    while year < 2001:
        if year % 4 == 0:
            days_in_month[2] = 29
            if year % 100 == 0 and year % 400 != 0: 
                days_in_month[2] = 28
        else:
            days_in_month[2] = 28

        if day > days_in_month[month]:
            day -= days_in_month[month]
            month += 1

        if month > 12:
            year += 1
            month = 1

        if day_of_week % 7 == 0:
            if day == 1 and year != 1900:
                counter += 1
            day_of_week = 0

        day += 1
        day_of_week += 1

    return counter

print('The number of Sundays that fell on the first of the month during the twentieth century:', solution())
```

## Problem 20: Factorial Digit Sum <a id="problem20"></a>

n! means n $\times$ (n − 1) $\times$ ... $\times$ 3 $\times$ 2 $\times$ 1

For example, 10! = 10 $\times$ 9 $\times$ ... $\times$ 3 $\times$ 2 $\times$ 1 = 3628800,\
and the sum of the digits in the number 10! is 3 + 6 + 2 + 8 + 8 + 0 + 0 = 27.

Find the sum of the digits in the number 100!

### Solution

-

```python
def solution():
    n = 100

    n_factorial = 1
    for i in range(1, n+1):
        n_factorial *= i

    total =  sum([int(x) for x in str(n_factorial)])

    return total

print('The sum of the digits in the number 100!:', solution())
```
